```python
import json

c3 = json.load(open('./cmip3variables.json'))
c5 = json.load(open('./cmip5variables.json'))
```

```python
c3s = []

dims = []
units = []


for table,val in c3.items():
    for vname,var in val.items():
        if 'standard_name' in var:
            ndim = '' 
            nunit = ''
            for i in var['dimensions'].split():
                
                if i not in dims:
                    dims.append(i)

                ndim += f'{dims.index(i)}&'
            
            for i in var['units'].split(): 
                if i not in units:
                    units.append(i)

                nunit += f'{units.index(i)}&'          




            dummy = f"{vname},{table},{var['standard_name']},{var['long_name']},{nunit},{ndim},{var.get('cell_methods','')},{var.get('valid_max','')}"
            c3s.append(dummy)

with open('cmip3compressed.csv','w') as f:
    f.write('\n'.join(c3s))
```
